## PRIME LEAGUE DATASET CLEANING

```python
from csv import reader
dataset = list(reader(open('football_data.csv', encoding='UTF-8')))
```

```python
dataset[:10]
```

```python
header = dataset[0]
dataset = dataset[1:]
```

```python
season_06 = [i for i in dataset if i[-1] == '2006-2007']
```

```python
[i[-1] for i in season_06]
```

```python
club_names = [i[0] for i in dataset]
unique_club_names = sorted(list(set(club_names)))
```

```python
table = [[i,0,0,0,0,0,0,0,0] for i in unique_club_names]
```

```python
table[:10]
```

```python
for i in season_06:
    home_team = i[0]
    away_team = i[1]
    home_goals = int(float(i[2]))
    away_goals = int(float(i[3]))
    result = i[4]
    
    if result == 'D':
            for j in table:
                if j[0] == home_team:
                    j[1] += 1 #match played
                    j[3] += 1 #draw match
                    j[-1] += 1 #points
                    j[5] += home_goals
                    j[6] += away_goals
                    j[7] = j[5] - j[6]
                if away_team == j[0]:
                    j[1] += 1 #match played
                    j[3] += 1 #draw match
                    j[-1] += 1 #points
                    j[5] += away_goals
                    j[6] += home_goals
                    j[7] = j[5] - j[6]
    elif result == 'H':
            for j in table:
                if j[0] == home_team:
                    j[1] += 1 #match played
                    j[2] += 1 #win match
                    j[-1] += 3 #points
                    j[5] += home_goals
                    j[6] += away_goals
                    j[7] = j[5] - j[6]
                if away_team == j[0]:
                    j[1] += 1 #match played
                    j[4] += 1 #loss match
                    j[5] += away_goals
                    j[6] += home_goals
                    j[7] = j[5] - j[6]
    elif result == 'A':
        for j in table:
            if j[0] == home_team:
                j[1] += 1 #match played
                j[4] += 1 #loss match
                j[5] += home_goals
                j[6] += away_goals
                j[7] = j[5] - j[6]
            if away_team == j[0]:
                j[1] += 1 #match played
                j[2] += 1 #win match
                j[5] += away_goals
                j[6] += home_goals
                j[7] = j[5] - j[6]
                j[-1] += 3
```

```python
table = sorted(table,key = lambda lst: lst[-1], reverse=True)
```

```python
table.insert(0,['Club', 'MP', 'W','D','L','GF','GA','GD','Pts'])
```
