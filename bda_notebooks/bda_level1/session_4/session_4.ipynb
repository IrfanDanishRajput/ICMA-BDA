```python
for item in artwork[1:]:
    gender = item[5]
    gender = gender.upper()
    if gender == '':
        item[5] = 'Gender Unknown/Other'
    item[5] = gender
    print(item[5])
```

```python
def clean_and_convert(date):
    if date != "":
        date = date.replace('-',"")
        date = int(date)
    return date
```

```python
for item in artwork[1:]:
    birth_date = item[3]
    death_date = item[4]
    
    birth_date = clean_and_convert(birth_date)
    death_date = clean_and_convert(death_date)
    
    item[3] = birth_date
    item[4] = death_date
```

```python
def process_date(value):
    bad_char = "-"
    if bad_char in value:
        value = value.split(bad_char)
        first_index = int(value[0])
        second_index = int(value[1])
        avg = (first_index + second_index) / 2
        value = avg
    else:
        value = int(value)
    return value
```
